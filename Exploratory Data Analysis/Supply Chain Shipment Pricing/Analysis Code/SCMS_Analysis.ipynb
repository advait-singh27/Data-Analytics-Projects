# Project Summary: Exploratory Data Analysis (EDA)

This project focuses on performing Exploratory Data Analysis (EDA) on the Supply Chain Shipment Pricing dataset using Python to understand the underlying structure, patterns, and quality of the data. The analysis aims to identify key characteristics such as data distributions, missing values, outliers, and relationships between variables before applying any modeling or advanced analytics.

Using libraries like Pandas, NumPy, Matplotlib, and Seaborn, the dataset is systematically explored through statistical summaries and visualizations. The process includes data inspection, data cleaning, univariate, bivariate, and multivariate analysis to uncover meaningful insights and potential trends.

The outcome of this project is a well-documented and structured EDA workflow that highlights important observations, supports data-driven decision-making, and prepares the dataset for further analysis or machine learning tasks. This project also demonstrates practical skills in data analysis, visualization, and analytical thinking, making it suitable for inclusion in a professional data analytics portfolio.

### Univariate Analysis

Univariate analysis examines one variable at a time to understand its distribution, frequency, and operational impact. This step helps identify dominant patterns, anomalies, and data quality issues before analyzing relationships between variables.

##### 1.Count of Shipments by Country:
This analysis evaluates the distribution of shipments across different countries to identify key operational regions. Countries with higher shipment volumes indicate major markets where Supply Chain operations may need increased capacity, workforce allocation, and infrastructure planning to ensure timely deliveries.

##### 2.Distribution of Shipment Modes:
Analyzing shipment modes such as Air, Sea, and Ground highlights how transportation choices are distributed. While air shipments offer faster delivery, they are used less frequently due to higher costs. Sea shipments, though slower, dominate the dataset, indicating a cost-driven logistics strategy within the Supply Chain.

##### 3.Average Freight Cost:
The average freight cost per shipment provides an overview of the cost structure. Higher average costs point to premium transport modes or heavier shipments, emphasizing the importance of optimizing route planning and mode selection in the Supply Chain.

##### 4.Line Item Quantity Distribution:
This analysis shows how shipment quantities are distributed. The majority of shipments contain smaller quantities, while large-volume shipments are relatively rare. These insights support decisions around order batching, inventory planning, and warehouse operations in the Supply Chain.

##### 5.Weight Distribution:
Shipment weight analysis reveals a wide range of delivery sizes, from lightweight parcels to heavy bulk shipments. Heavier shipments typically result in increased freight costs, underscoring the need for efficient handling and transport optimization for bulk goods within the Supply Chain.

##### 6.Vendors with Highest Orders:
Identifying vendors with the highest number of orders helps understand demand concentration. High-volume vendors represent strategic partners for the Supply Chain, where improved coordination and customized logistics solutions can enhance efficiency.




### Bivariate Analysis

Bivariate analysis explores the relationship between two variables, helping identify dependencies, correlations, and patterns that influence cost, delivery performance, and operational decisions.

##### 1. Shipment Mode vs Freight Cost
This analysis examines how freight costs vary across different shipment modes. Air shipments show significantly higher costs compared to sea and ground shipments, reinforcing the trade-off between speed and expense in Supply Chain operations.

##### 2.Shipment Weight vs Freight Cost
By analyzing shipment weight against freight cost, a positive relationship is observed where heavier shipments incur higher transportation costs. This insight supports weight-based pricing optimization and consolidation strategies within the Supply Chain.

##### 3.Country vs Shipment Mode
Studying shipment modes across countries reveals regional logistics preferences. Certain regions rely more on sea freight due to geographical advantages, while others depend on air or ground transport for faster turnaround times.

##### 4.Vendor vs Shipment Volume
Comparing vendors with shipment volumes helps identify high-impact partners. Vendors with consistently large shipment volumes may benefit from negotiated pricing, dedicated logistics routes, or service-level agreements in the Supply Chain.

### Multivariate Analysis

Multivariate analysis considers three or more variables simultaneously to uncover complex patterns that cannot be identified through univariate or bivariate analysis alone.

##### 1.Shipment Mode, Weight, and Freight Cost
This analysis reveals how shipment mode and weight together influence freight costs. For example, heavy air shipments incur disproportionately higher costs, making them key candidates for optimization within the Supply Chain.

##### 2.Country, Vendor, and Shipment Volume
Examining shipment volume across vendors and countries highlights regional vendor dependencies. These insights support strategic decisions related to vendor diversification and regional logistics planning.

##### 3.Quantity, Weight, and Delivery Cost
Analyzing line item quantity alongside weight and freight cost helps determine whether higher quantities lead to cost efficiency or increased expenses, guiding inventory and shipment consolidation strategies in the Supply Chain.

#### Data Wrangling

```python
df['Shipment Mode'].fillna('Unknown', inplace=True)
df['Dosage'].fillna('Not Applicable', inplace=True)
df['Line Item Insurance (USD)'].fillna(0, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
#Converting Numeric Columns
df['Weight (Kilograms)'] = pd.to_numeric(
    df['Weight (Kilograms)'], errors='coerce'
)

df['Freight Cost (USD)'] = pd.to_numeric(
    df['Freight Cost (USD)'], errors='coerce'
)
```

```python
#Converting Date Columns
date_cols = [
    'PQ First Sent to Client Date',
    'PO Sent to Vendor Date',
    'Scheduled Delivery Date',
    'Delivered to Client Date',
    'Delivery Recorded Date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Check the cleaned data
print(df.info())
```

```python
df.isnull().sum()
```

- Date-related missing values represent incomplete shipment lifecycle stages and were retained.
- Numeric missing values in weight and freight cost resulted from type conversion and reflect unavailable data.
- No rows were dropped to avoid data loss.
- No artificial imputation was performed to preserve analytical integrity.


#### Univariate Analysis

```python
#Count of Shipments by Country
plt.figure(figsize=(10,6))
sns.countplot(
    y='Country',
    data=df,
    order=df['Country'].value_counts().index
)
plt.title('Count of Shipments by Country')
plt.xlabel('Number of Shipments')
plt.ylabel('Country')
plt.show()
```

- Shipments are concentrated in a limited number of countries.
- These countries represent key operational regions in the supply chain.

```python
#Distribution of Shipment Modes
valid_modes = ['Air', 'Air Charter', 'Truck', 'Ocean']

plot_df = df[df['Shipment Mode'].isin(valid_modes)]


plt.figure(figsize=(6,4))
sns.countplot(
    x='Shipment Mode',
    data=plot_df,
    order=valid_modes
)
plt.title('Distribution of Shipment Modes')
plt.xlabel('Shipment Mode')
plt.ylabel('Count')
plt.show()
```

- Air is the most frequently used shipment mode in the dataset, indicating a strong preference for faster delivery despite higher costs.
- Ocean and Truck modes are used less frequently, likely for cost-efficient or region-specific shipments.
- Air Charter appears in a limited number of cases, suggesting its use for specialized or urgent deliveries.
- Note: Shipment mode values labeled as "Unknown" were excluded from this visualization to focus on valid transportation modes. The underlying dataset was not altered, ensuring data integrity while improving interpretability of the analysis.

```python
#Freight Cost (USD) Distribution
plt.figure(figsize=(8,5))
sns.histplot(
    df['Freight Cost (USD)'],
    bins=30,
    kde=True
)
plt.title('Freight Cost Distribution')
plt.xlabel('Freight Cost (USD)')
plt.ylabel('Frequency')
plt.show()
```

- Freight cost distribution is right-skewed.
- A small number of shipments incur significantly higher transportation costs.

```python
#Weight (Kilograms) Distribution
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Weight (Kilograms)'])
plt.title('Weight Distribution')
plt.xlabel('Weight (Kilograms)')
plt.show()
```

- Most shipments fall within a lower weight range.
- Heavier shipments appear as outliers and may contribute disproportionately to logistics costs.

```python
#Line Item Quantity Distribution
plt.figure(figsize=(8,5))
sns.histplot(
    df['Line Item Quantity'],
    bins=30
)
plt.title('Line Item Quantity Distribution')
plt.xlabel('Line Item Quantity')
plt.ylabel('Frequency')
plt.show()
```

- The majority of shipments consist of smaller quantities.
- High-quantity shipments are less frequent but operationally significant.

```python
#Top 10 Vendors by Shipment Volume
top_vendors = df['Vendor'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(
    x=top_vendors.values,
    y=top_vendors.index
)
plt.title('Top 10 Vendors by Shipment Volume')
plt.xlabel('Number of Shipments')
plt.ylabel('Vendor')
plt.show()
```

- Shipment volume is highly concentrated among a small number of vendors.
- These vendors are key partners within the supply chain.

#### Bivariate Analysis

```python
#Shipment Mode vs Freight Cost
valid_modes = ['Air', 'Air Charter', 'Truck', 'Ocean']

mode_cost_df = df[
    (df['Shipment Mode'].isin(valid_modes)) &
    (df['Freight Cost (USD)'].notna())
]

plt.figure(figsize=(8,5))
sns.boxplot(
    x='Shipment Mode',
    y='Freight Cost (USD)',
    data=mode_cost_df,
    order=valid_modes
)
plt.title('Shipment Mode vs Freight Cost')
plt.xlabel('Shipment Mode')
plt.ylabel('Freight Cost (USD)')
plt.show()
```

- Air shipments have the highest median freight cost, reflecting the premium associated with faster delivery.
- Ocean and Truck modes show lower and more stable freight costs, making them cost-efficient alternatives.
- Air Charter shipments exhibit higher cost variability, indicating use in exceptional or urgent cases.
- Note:
Records with "Unknown" shipment mode and missing freight cost values are excluded only for this analysis to improve interpretability.

```python
#Weight (Kilograms) vs Freight Cost (USD)
weight_cost_df = df[
    df['Weight (Kilograms)'].notna() &
    df['Freight Cost (USD)'].notna()
]

plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Weight (Kilograms)',
    y='Freight Cost (USD)',
    data=weight_cost_df,
    alpha=0.6
)
plt.title('Weight vs Freight Cost')
plt.xlabel('Weight (Kilograms)')
plt.ylabel('Freight Cost (USD)')
plt.show()
```

- Freight cost increases as shipment weight increases, indicating a positive relationship.
- Heavier shipments contribute disproportionately to overall transportation costs.
- The spread suggests that shipment mode and route may further influence cost beyond weight alone.

```python
# Select top 10 countries by shipment count
top_countries = df['Country'].value_counts().head(10).index

filtered_df = df[
    (df['Country'].isin(top_countries)) &
    (df['Shipment Mode'].isin(valid_modes))
]

plt.figure(figsize=(12,6))
sns.countplot(
    x='Country',
    hue='Shipment Mode',
    data=filtered_df
)
plt.title('Shipment Mode Distribution for Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Shipments')
plt.xticks(rotation=45)
plt.legend(title='Shipment Mode')
plt.show()
```

- Shipment mode usage varies significantly across top countries.
- Air shipments dominate in several high-volume countries, indicating time-sensitive deliveries.
- Ocean and Truck modes are more prevalent in specific regions, reflecting cost and geographic considerations.

```python
#Vendor vs Shipment Volume
top_vendors = df['Vendor'].value_counts().head(5).index

vendor_df = df[df['Vendor'].isin(top_vendors)]

plt.figure(figsize=(10,5))
sns.countplot(
    y='Vendor',
    data=vendor_df,
    order=top_vendors
)
plt.title('Shipment Volume for Top Vendors')
plt.xlabel('Number of Shipments')
plt.ylabel('Vendor')
plt.show()
```

- Shipment volume is highly concentrated among a small number of vendors.
- These vendors represent key supply chain partners and are critical to operational continuity.

#### Multivariate Analysis

```python
#Shipment Mode + Weight + Freight Cost
valid_modes = ['Air', 'Air Charter', 'Truck', 'Ocean']

multi_df = df[
    (df['Shipment Mode'].isin(valid_modes)) &
    (df['Weight (Kilograms)'].notna()) &
    (df['Freight Cost (USD)'].notna())
]

plt.figure(figsize=(10,6))
sns.scatterplot(
    x='Weight (Kilograms)',
    y='Freight Cost (USD)',
    hue='Shipment Mode',
    data=multi_df,
    alpha=0.6
)
plt.title('Weight vs Freight Cost by Shipment Mode')
plt.xlabel('Weight (Kilograms)')
plt.ylabel('Freight Cost (USD)')
plt.legend(title='Shipment Mode')
plt.show()
```

- Freight cost increases with shipment weight across all shipment modes.
- Air and Air Charter shipments show significantly higher freight costs for similar weights.
- Truck and Ocean shipments demonstrate more cost-efficient behavior for heavier shipments.

```python
#Country + Shipment Mode + Freight Cost
top_countries = df['Country'].value_counts().head(5).index

country_mode_cost_df = df[
    (df['Country'].isin(top_countries)) &
    (df['Shipment Mode'].isin(valid_modes)) &
    (df['Freight Cost (USD)'].notna())
]

plt.figure(figsize=(12,6))
sns.boxplot(
    x='Country',
    y='Freight Cost (USD)',
    hue='Shipment Mode',
    data=country_mode_cost_df
)
plt.title('Freight Cost by Country and Shipment Mode')
plt.xlabel('Country')
plt.ylabel('Freight Cost (USD)')
plt.xticks(rotation=45)
plt.legend(title='Shipment Mode')
plt.show()
```

- Freight cost varies significantly across countries depending on shipment mode.
- Air shipments consistently incur higher costs across all regions.
- Ocean and Truck modes show lower and more stable freight costs, particularly in high-volume countries.

```python
#Vendor + Shipment Mode + Shipment Volume
top_vendors = df['Vendor'].value_counts().head(5).index

vendor_mode_df = df[
    (df['Vendor'].isin(top_vendors)) &
    (df['Shipment Mode'].isin(valid_modes))
]

plt.figure(figsize=(12,6))
sns.countplot(
    x='Vendor',
    hue='Shipment Mode',
    data=vendor_mode_df
)
plt.title('Shipment Mode Distribution for Top Vendors')
plt.xlabel('Vendor')
plt.ylabel('Number of Shipments')
plt.xticks(rotation=45)
plt.legend(title='Shipment Mode')
plt.show()
```

- Shipment mode usage varies significantly across vendors.
- Some vendors rely heavily on air transport, indicating time-sensitive supply commitments.
- Other vendors make greater use of ocean or truck shipments, suggesting cost-optimized logistics strategies.

```python
#Quantity + Weight + Freight Cost
qty_weight_cost_df = df[
    (df['Line Item Quantity'].notna()) &
    (df['Weight (Kilograms)'].notna()) &
    (df['Freight Cost (USD)'].notna())
]

plt.figure(figsize=(10,6))
sns.scatterplot(
    x='Line Item Quantity',
    y='Freight Cost (USD)',
    size='Weight (Kilograms)',
    data=qty_weight_cost_df,
    alpha=0.6,
    legend=False
)
plt.title('Freight Cost vs Quantity (Bubble Size = Weight)')
plt.xlabel('Line Item Quantity')
plt.ylabel('Freight Cost (USD)')
plt.show()
```

- Higher quantities do not always result in lower freight costs.
- Shipment weight plays a more significant role than quantity in determining freight cost.
- Consolidation strategies should prioritize weight optimization rather than quantity alone.

## 🔍 Key Insights from Exploratory Data Analysis

Based on univariate, bivariate, and multivariate analysis of the supply chain delivery dataset, the following key insights were derived:

### 1️⃣ Shipment Mode Utilization
- Air transport is the most frequently used shipment mode, indicating a strong emphasis on faster delivery timelines.
- Ocean and Truck modes are used less frequently and primarily in scenarios where cost efficiency or geographical feasibility is prioritized.
- Air Charter shipments, while limited in volume, are associated with specialized or urgent delivery requirements.
- Missing shipment modes were preserved as `"Unknown"` during preprocessing to maintain data integrity and excluded only during visualization for clarity.

### 2️⃣ Freight Cost Drivers
- Freight cost is heavily influenced by shipment mode and shipment weight.
- Air and Air Charter shipments consistently incur higher freight costs across all regions and vendors.
- Ocean and Truck shipments demonstrate lower and more stable cost patterns, making them suitable for bulk or non-urgent deliveries.

### 3️⃣ Impact of Weight and Quantity
- Shipment weight has a stronger impact on freight cost than line item quantity.
- Heavier shipments contribute disproportionately to transportation costs, regardless of the number of items shipped.
- Increasing quantity alone does not guarantee cost efficiency unless accompanied by optimized weight distribution.

### 4️⃣ Regional Shipment Patterns
- Shipment mode preferences vary significantly by country.
- Certain high-volume countries rely predominantly on air shipments, suggesting time-sensitive or critical supply needs.
- Other regions show greater usage of ocean or truck modes, reflecting cost-driven logistics strategies.

### 5️⃣ Vendor-Level Insights
- Shipment volume is highly concentrated among a small number of vendors.
- Vendors exhibit distinct shipment mode preferences, indicating vendor-specific logistics strategies.
- High-dependency vendors represent key partners and potential candidates for negotiated freight contracts or optimized routing.

### 6️⃣ Data Quality & Lifecycle Observations
- Missing date values in PQ and PO stages indicate incomplete shipment lifecycle progression, reflecting real operational workflows rather than data errors.
- Missing weight and freight cost values highlight gaps in data capture and present opportunities for improving operational reporting.

## 📌 Business Recommendations to Achieve Objectives

Based on the insights derived from the exploratory data analysis, the following recommendations can help the client optimize supply chain performance, reduce logistics costs, and improve overall operational efficiency:

### 🔹 Optimize Shipment Mode Selection
- Air shipments dominate the supply chain but incur significantly higher freight costs.
- For non-urgent deliveries, shifting shipments from air to **ocean or truck transport** can result in substantial cost savings without compromising service levels.

- The client should focus on:
  - Improved packaging design  
  - Load consolidation  
  - Optimized routing  
- These measures are especially important for high-cost air shipments.

### 🔹 Adopt Strategic Vendor Management
- A small group of vendors accounts for a large proportion of shipment volume.
- Prioritizing these vendors for:

  - Volume-based pricing discounts  
  - Collaborative logistics planning

### 🔹 Improve Data Capture and Visibility
- Missing values in weight, freight cost, and shipment lifecycle dates indicate gaps in operational reporting.
- Strengthening data capture processes will:
  - Improve cost transparency

  - Support reliable future analytics and decision-making

### 🔹 Use Region-Specific Logistics Strategies
- Shipment mode preferences vary by country due to geography, infrastructure, and urgency.
- The client should adopt **region-specific logistics strategies** that balance speed and cost while accounting for regulatory and operational constraints.
