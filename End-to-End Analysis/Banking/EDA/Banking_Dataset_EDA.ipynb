## Exploratory Data Analysis – Banking Dataset

The objective of this notebook is to perform exploratory data analysis (EDA) on a banking customer dataset to understand customer demographics, financial behavior, and product usage. The insights derived here will guide feature engineering and support the creation of an interactive Power BI dashboard.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Banking.csv')
df.head()
```

```python
# Check the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Get a concise summary of the DataFrame
print("\nDataFrame Info:")
df.info()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
df["Estimated Income"]
```

```python
# Define income band boundaries
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']

# Create the 'Income Band' column using pd.cut
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, include_lowest=True)
```

```python
# Examine the distribution of unique categories in categorical columns
categorical_cols = df[["Risk Weighting","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Occupation","Income Band"]].columns
for col in categorical_cols:
  # if col in ["Client ID","Name","Joined Bank"]:
  #   continue
  print(f"\nValue Counts for '{col}':")
  display(df[col].value_counts())
```

```python
# Generate descriptive statistics for numerical columns
print("\nDescriptive Statistics for Numerical Columns:")
display(df.describe())
```

```python
df['Joined Bank'] = pd.to_datetime(df['Joined Bank'], format='%d-%m-%Y')
print(df['Joined Bank'].dtype)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Numerical analysis and exploration
numerical_cols = ['Fee Structure','Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Select numerical columns for correlation analysis
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
pairs_to_plot = [
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
    ('Checking Accounts', 'Foreign Currency Account'),
    ('Age', 'Superannuation Savings'),
    ('Estimated Income', 'Checking Accounts'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans'),
]


for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(5, 4))
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha': 0.4},     # semi-transparent points
        line_kws={'color': 'yellow'}       # best-fit line color
    )
    plt.title(f'Relationship between {x_col} and {y_col}', fontsize=10)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()
```

## 🔍 Final Insights from Exploratory Data Analysis

Based on the exploratory data analysis conducted on the banking customer dataset, the following key insights were identified. These insights provide a clear understanding of customer financial behavior and serve as a foundation for dashboard development.

---

### 1️⃣ Customer Demographics & Income Distribution
- The customer base spans a wide age range, with a higher concentration in the middle-aged segment.
- Estimated income shows a right-skewed distribution, indicating that while most customers fall into low-to-mid income ranges, a smaller group of high-income customers contributes significantly to overall wealth.
- Age and income do not exhibit a strong linear relationship, suggesting income growth is influenced by factors beyond age alone.

---

### 2️⃣ Asset Holding Behavior
- Bank deposits, checking accounts, and savings accounts show positive relationships, indicating that customers who actively engage with one savings product tend to hold balances across multiple account types.
- Customers with higher superannuation savings generally belong to older age groups, reflecting long-term financial planning behavior.
- Foreign currency accounts are held by a smaller subset of customers, likely representing higher-value or internationally active clients.

---

### 3️⃣ Liability and Credit Usage Patterns
- A positive relationship exists between bank loans and credit card balances, suggesting that customers with higher borrowing capacity tend to utilize multiple credit products.
- Business lending is strongly associated with higher bank loan values, indicating that business-oriented customers represent a higher credit exposure segment.
- The number of credit cards held does not directly correlate with higher balances, implying responsible credit usage among most customers.

---

### 4️⃣ Income vs Product Engagement
- Customers with higher estimated income generally maintain higher balances in checking and savings accounts.
- However, higher income does not always translate into proportionally higher borrowing, indicating varying risk appetites and financial strategies.
- Income-driven segmentation can effectively distinguish high-value customers for targeted services.

---

### 5️⃣ Risk Weighting Insights
- Risk weighting varies across customers with similar income levels, highlighting that income alone is not sufficient for risk assessment.
- Customers with higher liabilities tend to exhibit higher risk weighting, reinforcing the importance of combining asset and liability views for credit risk evaluation.

---

### 6️⃣ Loyalty and Fee Structure Patterns
- Higher loyalty classifications are generally associated with stronger asset positions and more diversified product usage.
- Fee structure categories show distinct customer distribution patterns, suggesting potential opportunities for personalized pricing and loyalty-based incentives.

---

## 📊 Key Takeaways for Dashboard Design

- **Net worth, total assets, and total liabilities** should be core KPIs.
- Income, risk weighting, and loyalty classification are effective **slicers** for customer segmentation.
- Asset vs liability relationships provide strong storytelling visuals for business and risk stakeholders.
- Product penetration views (loans, deposits, cards) enable cross-sell and upsell analysis.

---

## ✅ Conclusion

The EDA reveals clear patterns in customer financial behavior, highlighting the importance of viewing customers holistically across income, assets, liabilities, and risk. These insights enable data-driven decision-making and provide a strong analytical foundation for building an interactive Power BI dashboard focused on customer value, risk assessment, and product strategy.
